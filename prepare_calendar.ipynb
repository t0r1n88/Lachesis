```python
"""
скрипт для обработки календяра проф мероприятий
"""
```

```python
import pandas as pd
import re
```

```python
def extract_quantity(row):
    """
    функция для извлечения значений
    """
    data = row[6:]
    
    itog = 0
    for value in data:
        result = re.findall(r'(\d+)\s*(?:уче|чел|шк|ч)',value) # ищем числа
        if result:
            _lst = list(map(int,result))
            itog +=sum(_lst)
        else:
            itog += 0
    return itog
```

```python
df2022 = pd.read_excel('data/Календарь профориентационных мероприятий 2022.xlsx',skiprows=1,dtype=str)
```

```python
df2022 = df2022.dropna(subset=['Наименование ПОО']) # очищаем от пустых строк в списке названий ПОО
```

```python
df2022.fillna('0',inplace=True)
```

```python
# очищаем датафрейм от лишних колонок

url_lst = [value for value in df2022.columns if 'Ссылка' in value]

df2022.drop(columns=url_lst,inplace=True)
```

```python
df2022['Результат'] = df2022.apply(extract_quantity,axis=1)
```

```python
out_df2022=df2022.drop(columns=df2022.columns[5:-1])
```

```python
out_df2022.drop(columns=out_df2022.columns[1:-1],inplace=True)
```

```python
out_df2022=out_df2022.groupby(by='Наименование ПОО').sum() # суммируем
out_df2022.sort_values(by='Результат',ascending=False,inplace=True)
out_df2022.to_excel('Количество участников за 2022.xlsx')
```

```python
df2023 = pd.read_excel('data/Календарь профориентационных мероприятий 2023.xlsx',skiprows=1,dtype=str)


df2023 = df2023.dropna(subset=['Наименование ПОО']) # очищаем от пустых строк в списке названий ПОО

df2023.fillna('0',inplace=True)

# очищаем датафрейм от лишних колонок

url_lst = [value for value in df2023.columns if 'Ссылка' in value]

df2023.drop(columns=url_lst,inplace=True)

df2023['Результат'] = df2023.apply(extract_quantity,axis=1)

out_df2023=df2023.drop(columns=df2023.columns[5:-1])

out_df2023.drop(columns=out_df2023.columns[1:-1],inplace=True)

out_df2023=out_df2023.groupby(by='Наименование ПОО').sum() # суммируем
out_df2023.sort_values(by='Результат',ascending=False,inplace=True)
out_df2023.to_excel('Количество участников за 2023.xlsx')
```
